```python
import pandas as pd

# URLs de las bases de datos (Copia este bloque completo)
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargamos los datos
tienda = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

print("¡Listo! Datos cargados sin errores.")
```

```python
# Sumamos la columna 'Precio' para cada una de las tiendas
fact_t1 = tienda['Precio'].sum()
fact_t2 = tienda2['Precio'].sum()
fact_t3 = tienda3['Precio'].sum()
fact_t4 = tienda4['Precio'].sum()

print(f"Facturación Tienda 1: ${fact_t1:,.2f}")
print(f"Facturación Tienda 2: ${fact_t2:,.2f}")
print(f"Facturación Tienda 3: ${fact_t3:,.2f}")
print(f"Facturación Tienda 4: ${fact_t4:,.2f}")
```

```python
# Analizamos la frecuencia de las categorías
tiendas_lista = [tienda, tienda2, tienda3, tienda4]

for i, df in enumerate(tiendas_lista, 1):
    categoria_top = df['Categoría del Producto'].value_counts().idxmax()
    ventas = df['Categoría del Producto'].value_counts().max()
    print(f"Tienda {i} - Categoría líder: {categoria_top} ({ventas} ventas)")
```

```python
# Calculamos la media de la columna 'Calificación'
for i, df in enumerate(tiendas_lista, 1):
    promedio = df['Calificación'].mean()
    print(f"Tienda {i} - Calificación promedio: {promedio:.2f} estrellas")
```

```python
for i, df in enumerate(tiendas_lista, 1):
    top_prod = df['Producto'].value_counts().idxmax()
    bottom_prod = df['Producto'].value_counts().idxmin()
    print(f"Tienda {i}: [TOP: {top_prod}] | [MENOS VENDIDO: {bottom_prod}]")
```

```python
# Promedio de la columna 'Costo de envío'
for i, df in enumerate(tiendas_lista, 1):
    envio = df['Costo de envío'].mean()
    print(f"Tienda {i} - Costo envío promedio: ${envio:,.2f}")
```
